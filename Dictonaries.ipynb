```python
 x={}   # empty dic.
 print(type(x))
```

```python
# Create dict. by passing values
x={"v1":"Apple","v2":"Mango"}
print(x)
```

```python
# using dict. keyword
a=dict(v1="Apple",v2="Mango")
print(a)
```

```python
# Access dict. elements
d={"v1":10,"v2":20,"v3":30,"v4":40}
print(d["v1"],d["v2"],d["v3"])
a=d.get("v5","unknown")  #If the value user enter doesn't exixt
print(a)
```

```python
# Modify dict. elements
#Add elements in a dict.
d={"v1":10,"v2":20,"v3":30,"v4":40}
d["v5"]=50
print(d)
```

```python
#remove elements from a dict.
d={"v1":10,"v2":20,"v3":30,"v4":40}
del d["v1"]  #always remove with the key not by value of the key
print(d)
```

```python
a=d.pop("v1")
print(a)   #10
```

```python
d.pop("v1")
print(d)  #{'v2': 20, 'v3': 30, 'v4': 40}
```

```python
# Methods
#get() method
d={"v1":10,"v2":20,"v3":30,"v4":40}
print(d.get("v1"))
```

```python
#keys() method
print(d.keys())
print(d.values())
```

```python
#update() method
d.update({"v2":25,"v5":50})
print(d)
```

```python
#clear() method
d.clear()
print(d)
```
