To find reverse of a number and check whether the number is palindrome or not.

```python
# Take input from the user
a=int(input("enter a number:"))
rev=0
# Initialize reverse variable to store reversed number
temp=a
# Store original number in a temporary variable
# Loop until the number becomes 0
while a>0:
    digit=a%10              # Extract the last digit
    rev = rev * 10 + digit  # Build the reversed number
    print(digit, end="")    # Print digits in reverse order
    a = a // 10             # Remove the last digit
    # Check whether the original number and reversed number are equal
if temp==rev:
  print("palindrome number:")
else:
  print("not palindrome")
```

Program to check whether a number is an Armstrong number


```python
 # Take input from the user
num=int(input())
original = num
# Store the original number
sum=0
#Initialize sum variable
# Loop through each digit of the number
while num>0:
    digit=num%10  # Extract the last digit
    sum+=digit**3 # Raise digit to power of total digits and add to sum
    num=num//10   # Remove the last digit
    # Check if the calculated sum is equal to the original number
if sum==original:
    print("Armstrong number")
else:
    print("Not an armstrong number")
```
