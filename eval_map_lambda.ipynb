```python
# eval()
a=eval("10*3+10-20") #enter every value in string inside eval()
print(a)  # 20
```

```python
x=5
a=eval("x*5")
print(a)
```

```python
a="10*10"
print(eval(a))
```

```python
# map() #### map(function, iterable)
# double each number
n=[1,2,3,4,5]
double=list(map(lambda x:x*2,n))  #for each x it(lambda) returns x*2
print(double)
```

```python
# #convert string to integer
a=["10", "20", "30"]
int=list(map(int, a))
print(int)
```

```python
# map over two list
a=[2,4,6]
b=[1,3,5]
sum=list(map(lambda x,y:x+y, a,b))
print(sum)
```

```python
# using a named function with map
def square(n):
    return n**2
squares=list(map(square, range(1,6)))
print(squares)
```

```python
# Lambda(Imp.)
# syntax: lambda args:expression
# basic lambda
square=lambda x:x**2
print(square(5))
```

```python
# multiple arguments
add =lambda x,y:x+y
print(add(2,5))
```

```python
# conditional logic
a=lambda n:"even" if n%2==0 else "odd"
print(a(8))
print(a(3))
print(a(4))
```

```python
# lambda with filter()
n=[1,2,3,4,5,6,7,8]
even=list(filter(lambda x:x%2==0, n))
print(even)
```

```python
# lambda with filter()
n=[1,2,3,4,5,6,7,8]
odd=list(filter(lambda x:x%2!=0, n))
print(odd)
```
