**Create tuple**

```python
a=() #empty tuple
print(type(a))
len(a)
```

```python
a=tuple([2,4,6,8,10]) #type conversion
print(a)
len(a)
```

```python
# Add elements in a tuple
tuple1=(10,20)
tuple2=tuple1+(30,40)
tuple2
```

**Access element in tuples**

```python
x=(1,3,5,7)
print(x[1])
print(x[-1])
print(x[-2::])
print(x[:2:2])
```

**Operations on tuples**
*   concatination
*   Duplicate
*   Validity

```python
# concatination
t1=(1,2,3,4,5)
t2=(6,7,8,9,10)
t3=t1+t2
print(t3)
```

```python
# duplicate
repeated=t3*2
print(repeated)
```

```python
repeated=t3*3
print(repeated)
```

```python
# Validity
print(1 in t3) #True
print(45 in t3)  #False
```

Packing and unpacking of tuples

```python
#packing
x=1,2,3,4,5
print(x)
print(type(x))
```

```python
# unpacking
a=(2,4,6,8)
a1,a2,a3,a4=a
print(a1)
print(a2)
print(a3)
print(a4)
```

```python
# Nested unpacking
x=(1,2,(3,4),5,6)
print(x[2])
```

```python
print(x[2][0])
```

```python
x=(1,2,(3,(4,5),6))
print(x[2])
```

```python
print(x[2][1])
```

```python
print(x[2][1][1])
```
