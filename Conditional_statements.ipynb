To check weather the person can vote or not.

```python
a=int(input()) #ask user to enter a number.
if a>=18: #check the condition is true or not.
    print("The person can vote.")
else:
    print("The person can not vote.")
```

To check student's grades on the basis of marks.

```python
a=int(input("Enter student marks:")) #ask the user to enter student marks.
if a>=90:
    print("Grade A") #if marks are 90 or more.
elif a>=80:
    print("Grade B") #if marks are 80-89.
elif a>65:
    print("Grade C") #if marks are 65-79.
elif a>50:
    print("PASS")    #if marks are 50-64.
else:
    print("Fail")    #if marks are below 50.
```

To check the given number is positive, negative or zero.

```python
a=int(input("enter a number:")) #ask the user to enter a number.
if a>0:
    print("The number is positive:") #check whether the number is positive.
elif a<0:
    print("The number is negative")  #check whether the number is negative.
else:
    print("The number is zero")      #check whether the number is zero.
```

To check which number is greater.

```python
#Take two numbers from user and print the greater number.
a=int(input("enter first number"))
b=int(input("enter second number"))
if a/b>1: #check whether the remainder is greater than 0.
    print("a is greater")
elif a/b==1: #check whether the remainder is equal to 0.
    print("a and b are equal")
else:    #check the remainder is less than 0.
    print("b is greater")
```
