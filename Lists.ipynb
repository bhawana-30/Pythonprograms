```python
# List/accessing element
Icecreams = ["kesarpista", "Chokobar", "vanilla"]
print(Icecreams[1])
first_icecream = Icecreams[0]
print('first icecream:',first_icecream)
Icecreams[0]="strawberry"
print("updated icecreams:",Icecreams)
```

```python
x=[1,2,3,4]
first_number = x[0]
print('first number:',first_number)
x[3]=5
print(x)
```

```python
del x[3],x[2]
print(x)
```

```python
x.remove(3)
print(x)
```

```python
# combining list
u=[10,20,30]
v=[40,50,60]
w=u+v
print(w)
```

```python
# sorting list
x=[2,4,10,5,20,11]
x.sort()
print(x)
```

```python
x.sort(reverse=True)  #reverse the list
print(x)
```

```python
# add element at last
x.append(32)
print(x)
```

```python
#remove last element
x.pop()
print(x)
```

```python
# list comprihension
y=[x**2 for x in range(11)]
print(y)
```

```python
# list comprihension with condition
even=[x for x in range(11) if x%2==0]
print(even)
odd=[x for x in range(11) if x%2!=0]
print(odd)
```

```python
# list comprihension with string manipulation
Icecreams = ["kesarpista", "Chokobar", "vanilla"]
Icecreams_upper=[icecream.upper() for icecream in Icecreams]
print(Icecreams_upper)
```

```python
x=[1,2,3,4,5,6,7]
y=x[2:5]
print(y)
```

```python
y=x[0:8:3]     # x[start point:end point+1:incriment state]
print(y)
```

```python
# Reverse a list
y=x[::-1]
print(y)
```
